```python
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score, recall_score
```

```python
interaction_profiles = pd.read_csv("/content/drive/Shareddrives/5461 Project/data/Challenge_GIN_release_profile_17804library_181query.txt", delimiter="\t",index_col=0)
go_term_annotations = pd.read_csv("/content/drive/Shareddrives/5461 Project/data/ChallengeA_release_GO_BP_20_500_17804library_1106terms.txt", delimiter="\t",index_col=0)
```

```python
# Exclude the first 5000 genes
train_interaction_profiles = interaction_profiles.iloc[5000:]
train_go_term_annotations = go_term_annotations.iloc[5000:]


# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(train_interaction_profiles, train_go_term_annotations, test_size=0.2, random_state=42)

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
train_interaction_profiles_normalized = scaler.fit_transform(train_interaction_profiles)

X_train, X_test, y_train, y_test = train_test_split(train_interaction_profiles_normalized, train_go_term_annotations, test_size=0.2, random_state=42)
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Input(shape=(181,)),
    tf.keras.layers.Dense(512, activation='elu'),
    #tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(256, activation='elu'),
    #tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(1106, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=20, batch_size=64, validation_split=0.2)
```

```python
y_pred = model.predict(X_test)
y_pred_binary = np.where(y_pred > 0.5, 1, 0)  # Change to 0s and 1s for y_pred_binary

# Convert y_test to binary values (1s and 0s)
y_test_binary = y_test.replace(-1, 0)

precision = precision_score(y_test_binary, y_pred_binary, average='micro')
recall = recall_score(y_test_binary, y_pred_binary, average='micro')

print("Precision: ", precision)
print("Recall: ", recall)
```
